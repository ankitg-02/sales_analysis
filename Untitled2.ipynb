```python
import pandas as pd
import numpy as np
```

```python
#importing the csv dataset
sp=pd.read_csv(r"/content/Sample-sales-data-excel.csv")
```

```python
#converting the dataset to dataframe
sample=pd.DataFrame(sp)
```

```python
#checking the empty values in the dataset
sample.isnull().sum()
```

```python
#filling the empty values with NaN
sample.fillna(np.NaN)
```

```python
#statistical describation of dataset
sample.describe(include='all')
```

```python
#count of customers in each segment
count_conumer=0
for i in sample['Segment']:
  if i=='Consumer':
    count_conumer+=1
print("count of conumers=",count_conumer)
count_corporate=0
for j in sample['Segment']:
  if j=='Corporate':
    count_corporate+=1
print("count of corporates=",count_corporate)
count_home_office=0
for k in sample['Segment']:
  if k=='Consumer':
   count_home_office+=1
print("count of home offices=",count_home_office)
max_segment=max(count_conumer,count_corporate,count_home_office)
print("max of segment=",max_segment)
```
