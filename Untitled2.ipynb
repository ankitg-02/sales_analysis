```python
import pandas as pd
import numpy as np
```

```python
#importing the csv dataset
sp=pd.read_csv(r"/content/Sample-sales-data-excel.csv")
```

```python
#converting the dataset to dataframe
sample=pd.DataFrame(sp)
```

```python
#checking the empty values in the dataset
sample.isnull().sum()
```

```python
sample.info()
```

```python
d=sample.describe(include='all')
d
```

```python
#filling the empty values
sample['Postal Code'].replace('NaN',d['Postal Code'][4],inplace=True)
sample['Region'].replace('NaN',d['Region'][2],inplace=True)
sample['Product ID'].replace('NaN',d['Product ID'][4],inplace=True)
sample['Category'].replace('NaN',d['Category'][4],inplace=True)
sample['Sub-Category'].replace('NaN',d['Sub-Category'][4],inplace=True)
sample['Product Name'].replace('NaN',d['Product Name'][4],inplace=True)
sample['Sales'].replace('NaN',d['Sales'][2],inplace=True)
sample['Quantity'].replace('NaN',d['Quantity'][2],inplace=True)
sample['Discount'].replace('NaN',d['Discount'][2],inplace=True)
sample['Profit'].replace('NaN',d['Profit'][2],inplace=True)
```

```python
#statistical describation of dataset
sample.describe(include='all')
```

```python
for i in sample.columns:
  print(i,':','\n',sample[i].unique())
```

```python
#count of customers in each segment
count_conumer=0
for i in sample['Segment']:
  if i=='Consumer':
    count_conumer+=1
print("count of conumers=",count_conumer)
count_corporate=0
for j in sample['Segment']:
  if j=='Corporate':
    count_corporate+=1
print("count of corporates=",count_corporate)
count_home_office=0
for k in sample['Segment']:
  if k=='Consumer':
   count_home_office+=1
print("count of home offices=",count_home_office)
max_segment=max(count_conumer,count_corporate,count_home_office)
print("max of segment=",max_segment)
```

```python
sales_via_segment=sample.groupby("Segment")
```
